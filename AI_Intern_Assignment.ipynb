```python
from pprint import pprint
from Questgen import main
```

```python
qe= main.BoolQGen()
```

```python
file_path = f"{root_dir}/Big Mac Index.txt"
with open(file_path, 'r') as file:
    # Read the entire contents of the file
    text = file.read()
```

```python
text = ' '.join(text.split())
```

```python
payload = {
            "input_text": text
        }
```

```python
output1 = qe.predict_boolq(payload)
pprint (output1)
```

```python
text1 = str(output1['Boolean Questions'])
```

```python
qg = main.QGen()
```

```python
output2 = qg.predict_mcq(payload)
pprint (output2)
```

```python
text2=output2['questions']
```

```python
output3 = qg.predict_shortq(payload)
pprint (output3)
```

```python
text3 = output3['questions']
```

```python
answer = main.AnswerPredictor()
```
